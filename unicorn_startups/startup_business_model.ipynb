```python
import pandas as pd

# Function to classify business models based on industry
def classify_business_model(industry):
    if 'Consumer' in industry or 'Retail' in industry or 'Media' in industry:
        return 'B2C'
    elif 'Enterprise' in industry or 'Industrials' in industry or 'Financial' in industry:
        return 'B2B'
    elif 'Healthcare' in industry:
        return 'B2B/B2C'
    else:
        return 'Unknown'

# Apply the classification function
data['Business Model'] = data['Industry'].apply(classify_business_model)

# Display the updated dataset
print(data)
```

```python
import pandas as pd
data.to_csv('updated_data.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# Count the occurrences of each business model
business_model_counts = data['Business Model'].value_counts()

# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    business_model_counts,
    labels=business_model_counts.index,
    autopct='%1.1f%%',
    startangle=140,
    textprops={'fontsize': 12}
)
plt.title('Distribution of Business Models Among Unicorn Companies', fontsize=14)
plt.show()
```
